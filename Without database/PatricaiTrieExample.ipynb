```python
from PatriciaTrie import PatriciaTrie
```

Creating some users in our system

```python
user = {
    'Alice': '00010',
    'Bob': '10100',
    'Sally': '111111',
    'Darkhan': '010111'
}
```

```python
t = PatriciaTrie(simple_hash=True)

t.create([(user['Alice'], 100), (user['Bob'], 50)], '0000')
for key in t.info:
    print('node id:', key, t.info[key]['hash'])
t.draw()
```

```python
t.spend(user['Alice'], user['Bob'], 10, '1000')
for key in t.info:
    print('node id:', key, t.info[key]['hash'])
t.draw()
```

```python
t.spend(user['Alice'], user['Sally'], 40, '1001')
t.draw()
```

```python
t.spend(user['Bob'], user['Sally'], 50, '1002')
t.draw()
```

```python
t.spend(user['Sally'], user['Darkhan'], 80, '1010')
t.draw()
```
